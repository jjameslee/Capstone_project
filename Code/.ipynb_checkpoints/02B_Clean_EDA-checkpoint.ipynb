This function will trim the time columns into more readible string objects

```python
def clean_time(df, column):
    cleaned = []
    for time in df[column]:
        cleaned.append(time.split('.')[0])
        
    df.insert(0, 'clean_' + column, cleaned)
```

```python
clean_time(lafd, 'on_scene_time_gmt')
clean_time(lafd, 'en_route_time_gmt')
clean_time(lafd, 'time_of_dispatch_gmt')
clean_time(lafd, 'incident_creation_time_gmt')

lafd.drop(columns=['en_route_time_gmt', 'incident_creation_time_gmt', 'on_scene_time_gmt', 'time_of_dispatch_gmt'], inplace=True)
```

#### **LAFD Response Time Metrics Definitions:**
 
**Response Time:**
The time interval that starts when first contact is made (either through 911 or the fire dispatch center) and ends when the first respondents arrives on-scene. ***(turnout time + travel time)***
 
**Call Processing Time:**
The time interval that starts when the call‐taker is presented with the call and creates the incident in CAD and ends when a fire station is notified and provided dispatch instructions via the LAFD’s Fire Station Alerting System. ***(dispatch time - incident creation time)***
 
**Turnout Time:**
The time interval between the time of dispatch and the en-route (wheels rolling) time.  Both station alarm and en-route times are required to measure this for each unit that responds.  Turnout time is calculated for each unit dispatched to each incident. ***(en route time - dispatch time)***
 
**Travel Time:**
The time interval that begins when the respondent is en-route to the incident and ends upon arrival on scene.  This requires one valid en-route time and one valid on-scene time for the incident.  Travel time can differ considerably amongst stations. ***(on scene time - en route time)***

    * According to the LAFD website many factors may impact travel time:
         * traffic, topography, road width, public events and unspecified incident locations.

```python
lafd['call_process_time'] = lafd['time_of_dispatch_secs'] - lafd['incident_creation_time_secs']
lafd['turnout_time'] = lafd['en_route_time_secs'] - lafd['time_of_dispatch_secs']
lafd['travel_time'] = lafd['on_scene_time_secs'] - lafd['en_route_time_secs']
lafd['response_time'] = lafd['turnout_time'] + lafd['travel_time']
```

```python
lafd.head()
```

Using the 'Randomized Incident Number' feature which provides the year and quarter of the incident in the first 6 digits and then followed by another 6 random numbers (privacy law), we were able to add the year and quarter feature to the dataset.

```python
year = []
for random in lafd.randomized_incident_number.values:
    year.append(int(str(random)[0:4]))

lafd['year'] = year
```

```python
quarter = []
for random in lafd.randomized_incident_number.values:
    quarter.append(int(str(random)[5:6]))
    
lafd['quarter'] = quarter
```

#### LAFD requests made by different regions in the LA county:

Let's take a look at how the emergency requests compare when we look at the LA County regions.

```python
sf_valley = ['ARLETA NC', 'CANOGA PARK NC', 'CHATSWORTH NC', 'ENCINO NC', 'FOOTHILL TRAILS DISTRICT NC',
            'GRANADA HILLS NORTH NC', 'GREATER TOLUCA LAKE NC', 'LAKE BALBOA NC', 'MID-TOWN NORTH HOLLYWOOD NC',
            'MISSION HILLS NC', 'NORTH HILLS WEST NC', 'NORTH HOLLYWOOD NORTHEAST NC', 'NORTHRIDGE SOUTH NC',
            'NORTHRIDGE WEST', 'PACOIMA NC', 'PANORAMA CITY NC', 'PORTER RANCH NC', 'RESEDA NC', 'SHERMAN OAKS NC',
            'STUDIO CITY NC', 'SUNLAND-TUJUNGA NC', 'SYLMAR NC', 'TARZANA NC', 'WEST HILLS NC',
            'WINNETKA NC', 'WOODLAND HILLS-WARNER CENTER NC', 'VAN NUYS NC']

west_side = ['BEL AIR-BEVERLY CREST NC', 'MAR VISTA CC', 'PALMS NC', 'SOUTH ROBERTSON NC', 'VENICE NC',
            'WEST LOS ANGELES NC', 'WESTSIDE NC', 'WESTWOOD NC', 'NC WESTCHESTER/PLAYA']

hollywood = ['HOLLYWOOD HILLS WEST NC', 'CENTRAL HOLLYWOOD NC', 'HOLLYWOOD STUDIO DISTRICT NC', 
            'HOLLYWOOD UNITED NC']

south_central = ['CENTRAL ALAMEDA NC', 'COMMUNITY AND NEIGHBORS FOR NINTH DISTRICT UNITY (CANNDU)',
                'EMPOWERMENT CONGRESS SOUTHEAST AREA NDC', 'EMPOWERMENT CONGRESS NORTH AREA NDC',
                'EMPOWERMENT CONGRESS CENTRAL AREA NDC', 'EMPOWERMENT CONGRESS WEST AREA NDC', 'SOUTH CENTRAL NC',
                'UNITED NEIGHBORHOODS OF THE HISTORIC ARLINGTON HEIGHTS, WEST ADAMS, AND JEFFERSON PARK COMMUNITY',
                'VOICES OF 90037', 'WATTS NC']

harbor = ['CENTRAL SAN PEDRO NC', 'COASTAL SAN PEDRO NC', 'NORTHWEST SAN PEDRO NC', 'WILMINGTON NC', 
          'HARBOR GATEWAY SOUTH NC', 'HARBOR CITY NC']

downtown = ['DOWNTOWN LOS ANGELES', 'HISTORIC CULTURAL NC']

east_side = ['BOYLE HEIGHTS NC', 'LINCOLN HEIGHTS NC', 'LA-32 NC']

north_east = ['EAGLE ROCK NC', 'GLASSELL PARK NC', 'GREATER CYPRESS PARK NC', 'HISTORIC HIGHLAND PARK NC']

central = ['GREATER ECHO PARK ELYSIAN NC', 'GREATER WILSHIRE NC', 'LOS FELIZ NC', 'SILVER LAKE NC', 'MID CITY NC',
          'MID CITY WEST CC', 'PICO UNION NC', 'RAMPART VILLAGE NC', 'WESTLAKE SOUTH NC']


regions = []

for nc in lafd['nc_name']:
    if nc in sf_valley:
        regions.append('sf_valley')
    elif nc in west_side:
        regions.append('west_side')
    elif nc in hollywood:
        regions.append('hollywood')
    elif nc in south_central:
        regions.append('south_central')
    elif nc in harbor:
        regions.append('harbor')
    elif nc in downtown:
        regions.append('downtown')
    elif nc in east_side:
        regions.append('east_side')
    elif nc in north_east:
        regions.append('north_east')
    elif nc in central:
        regions.append('central')
    else:
        regions.append(nc)
        
lafd['regions'] = regions
```

We can see that North East, East side, and Harbor of the LA regions have the lowest number of LAFD emergency reports. On the other hand, San Fernando Valley and South Central have significantly higher number of incidents reported. Granted, the number of neighborhood councils within the SF valley and South Central regions are much more than the number of neighborhood councils within the East Side and North East regions. 

```python
west_and_hollywood = lafd[(lafd['regions'].str.contains('west_side')) | (lafd['regions'].str.contains('hollywood'))]
south_central = lafd[lafd['regions'].str.contains('south_central')]


westandhollcount = west_and_hollywood.groupby('year')['year'].count()
southcentralcount = south_central.groupby('year')['year'].count()

plt.figure(figsize=(15,10))
plt.subplot(2, 1, 1)
westandhollcount.plot(color = 'blue')


plt.ylabel('Number of incidents', color='red')
plt.xlabel('Incident timeframe', color='red')
plt.title('Total LAFD incident responses from 2013 to 2017 in Hollywood & West-side Regions', fontsize=16, color='red');

plt.subplot(2, 1, 2)
southcentralcount.plot(color = 'red')

plt.ylabel('Number of incidents', color='blue')
plt.xlabel('Incident timeframe', color='blue')
plt.title('Total LAFD incident responses from 2013 to 2017 in South Central Regions', fontsize=16, color='blue');

plt.tight_layout()
```

Overall, there is a higher number of incidents reported in the South Central regions than their more-affluent Hollywood and West side counterparts. From early 2013 to the beginning of 2016, both regions steadily increased the number of emergency LAFD incidents; however, as it continued to increase in the Hollywood and West side region of LA, it droppped a bit in the South Central region.
